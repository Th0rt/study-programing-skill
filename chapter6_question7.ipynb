```python
import matplotlib.pyplot as plt
from chapter6.question7 import simulate_multipule
from random import randint


results = simulate_multipule(trial_number=100, max_famiry_size=100)

keys = list(results.keys())
averages = [result["average"] for result in results.values()]

# グラフを書く
plt.plot(keys, averages)

# x軸のラベル
plt.xlabel("famiry_count")

# y軸のラベル
plt.ylabel("average_female_ratio")

# 表示する
plt.show()
```

```python
import matplotlib.pyplot as plt
from chapter6.question7 import simulate_multipule
from random import randint


results = simulate_multipule(trial_number=100, max_famiry_size=1000)

keys = list(results.keys())
standard_deviation = [result["standard_deviation"] for result in results.values()]

# グラフを書く
plt.plot(keys, standard_deviation)

# x軸のラベル
plt.xlabel("famiry_count")

# y軸のラベル
plt.ylabel("standard_deviation")

# 表示する
plt.show()
```
